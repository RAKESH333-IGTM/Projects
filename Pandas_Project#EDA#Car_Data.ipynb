# EDA(Exploratory Data Analysis)

  Exploratory data analysis is the first step of data analysis, It helps in understanding the data before building the models or decision makings.
  ##### GOAL OF EDA:-
  - To clean, Understand and Prepare the data for modelling or further analysis.
  ##### Purpose of EDA
  - Understanding the data variables patterns and relationships.
  - Detecting/finding errors or the missing values.
  - Understanding the data structure.
  - Discover Outliers and Unusual Observations.
  - Discovering null values/ Missing values.
  - Helps in making decision on which models or techniques should apply later.

##### Basic To Do's of EDA:-
- Viewing fist and last few rows:-- df.head(), df.tail()
- For checking data_types and structure :---   df.info()
- To know the basic stats like mean,median and standard deviation :-- df.describe()
- Finding missing data:-  df.isnull().sum()
- For Visual understanding of the data(Bar, Histogram, and Boxplot etc..):-  sns.histplot and sns.boxplot
- Correlation between the variables :-- df.corr()  and sns.heatmap()

##### Common Visualisation tools:-
- HISTOGRAM :- To check the distribution <br>
        [sns.histplot() or plt.hist()]
- BOXPLOT   :- To find out the outliers.<br>
        [sns.boxplot() or plt.boxplot() ]
- SCATTER PLOT :- To unerstand the relationship between the two variables.<br>
        [sns.scatterplot() or plt.scatter()]
- HEATMAP :- To visualise correlations.<br>
        [sns.heatmap()]
- BARPLOT :- To compare categories. <br>
        [sns.barplot or plt.bar()]
- LINEPLOT :- Shows trends overtime. <br>
        [sns.lineplot() or plt.plot()]
- PANDAS PLOT :- Quick plots from Dataframes. <br>
        [df.plot('kind = bar'), df.plot.hist()]

## TREATING OUTLIERS

```python
car_df.describe()
## .describe give the complete numerical/measures data, with this we can clearly undestand the outlies based on the difference between the mean,meadian and standard deviation.
```

```python
#Outliers = car_df[car_df['INCOME']<lower_bound | car_df['INCOME']>upper_bound]

def remove_outliers(col):
    sorted(col)
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 - 1.5*IQR
    return lower_bound,upper_bound


lr,ur = remove_outliers(car_df['INCOME'])
lr,ur
car_df['INCOME'] = np.where(car_df['INCOME']<lr,lr,car_df['INCOME'])
car_df['INCOME'] = np.where(car_df['INCOME']>ur,ur,car_df['INCOME'])

lr,ur = remove_outliers(car_df['MILES CLOCKED'])
lr,ur
car_df['MILES CLOCKED'] = np.where(car_df['MILES CLOCKED']<lr,lr,car_df['MILES CLOCKED'])
car_df['MILES CLOCKED'] = np.where(car_df['MILES CLOCKED']>ur,ur,car_df['MILES CLOCKED'])


```

```python
def outlier_treatment(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_bound = Q1-(1.5*IQR)
    upper_bound = Q3+(1.5*IQR)
    return lower_bound,upper_bound

lr,ur = outlier_treatment(car_df['TRAVEL TIME'])
lr,ur
car_df['TRAVEL TIME'] = np.where(car_df['TRAVEL TIME']>ur,ur,car_df['TRAVEL TIME'])
car_df['TRAVEL TIME'] = np.where(car_df['TRAVEL TIME']<lr,lr,car_df['TRAVEL TIME'])
    
```

## TREATING NULL VALUES

We can see the null values that are present in the data/in respective columns. There are various ways and techniques to treat he data, but that is completely depends on what type of data we are treating either it is numerical or categorical.

##### DROP MISSING VALUES:-
      - In this case we drop the missing values from those variables. In case there are very few missing values then we can drop those values.
##### IMPUTE THE MEAN VALUE:-
      - For numerical column, we can impute the missing values with the mean values. Before imputing, we need to check the presence of outliers.because while imputing the mean values, there should not present the outliers in the data.
##### IMPUTE THE MEDIAN VALUE:-
      - For numerical columns, we can replace the null values with medians. if there is any outliers present in the column, then it is advisable to impute the median values.
##### IMPUTE THE MODE VALUE:-
     - For the categorical columns, we can impute the mode values. that is with the frequent ones.

# CORRELATION
- Correlation is a statistical measure that describe the strength and direction of the relationship between the two variables.
- POSITIVE CORRELATION :- If one variable increase, then another variable also increase.  ex:- sales vs profit, height vs weight.
- NEGATIVE CORRELATION :- If one variable increses, then another variable decreases.      ex:- exercise vs body fat%, speed vs distance, Speed vs Time.
- NO CORRELATION :- If there is no relation/no consistent pattern between the variables.  ex:- car age vs sex, miles clocked vs gender, shoe size vs IQ.

# GROUPING AND SUMMERIZING
- Grouping and aggregation is one of the most commonly used operations in data analysis.Especially while doing EDA, Where we summerizeand compare the groups statastics which data is same/common.
- for ex:- In real estate data, we have to compare average sales of different regions. or comparing the total profit of two segments.
  #### Grouping analysis are having three segments:-
- SPLITTING :- Splitting the data into groups(ex:- categories,segments,regions,etc...)
- APPLYING  :- Applying a function to each group(ex:- mean or total sales of particular customer segment)
- COMBINING :- Combining the results into a data structure showing the summery stats.

  SYNTAX:--
           df.groupby().agg(sum,mean,median)
